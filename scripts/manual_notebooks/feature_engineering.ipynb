```python
import pandas as pd
```

```python
csv_path = '../../data/clean_insurance.csv'
df = pd.read_csv(csv_path)
```

```python
X = df.drop('charges', axis=1) #Dataframe without charges
Y = df['charges'] #Target variable Series
```

```python
X.sample(5)
```

```python
Y.head()
```

```python
#Creating Interaction Feature for Linear Models
X['bmi_smoker'] = X['bmi'] * X['smoker_yes']
X['age_smoker'] = X['age'] * X['smoker_yes']
X['age_bmi'] = X['age'] * X['bmi']
```

```python
X.head()
```

```python
#Scaling numeric feature (Standard Scalar)
from sklearn.preprocessing import StandardScaler
num_cols = ['age', 'bmi', 'children', 'bmi_smoker', 'age_smoker', 'age_bmi']
scaleStandard = StandardScaler()
X_scaled = X.copy()
X_scaled[num_cols] = scaleStandard.fit_transform(X[num_cols])
```

```python
X_scaled.head(5)
```

```python
#Export to final csv
df_final = X_scaled.copy()
df_final['charges'] = Y
df_final.to_csv('../data/insurance_scaled_final.csv', index=False)
```
